### Create a line plot of the Total years of experience vs the current salary(taking the median salary for each of the different experience years) of the year 2018.

### Now, create the above plot again and add 2 more line plots to the same graph, that display the Total years of experience vs the median Yearly brutto salary (without bonus and stocks) of the year 2019 and 2020.

```
se_1=data_19.groupby('Years of experience').median()['Yearly brutto salary (without bonus and stocks)'] #aggregating years of experience , with the median of brutto salary in 2019 data.
temp_2=pd.DataFrame(se_1) #Converting the above series into an pandas dataframe.
temp_2.reset_index(inplace=True)#Resetting the index for the above created dataframe.
yoe_2019=temp_2['Years of experience'].tolist() #Picking up the years of experience column alone and converting it into an list.
ybs_2019=temp_2['Yearly brutto salary (without bonus and stocks)'].tolist() #Picking up the yearly brutto salary column alone and converting it into an list.
```

```
se_2=data_20.groupby('Total years of experience').median()['Yearly brutto salary (without bonus and stocks) in EUR'] #aggregating years of experience , with the median of brutto salary in 2020 data.
temp_3=pd.DataFrame(se_2) #Converting the above series into an pandas dataframe.
temp_3.reset_index(inplace=True) #Resetting the index for the above created dataframe.
yoe_2020=temp_3['Total years of experience'].tolist() #Picking up the total years of experience column alone and converting it into an list.
ybs_2020=temp_3['Yearly brutto salary (without bonus and stocks) in EUR'].tolist()  #Picking up the yearly brutto salary column alone and converting it into an list.
```

```
from plotly.subplots import make_subplots #Imprting plotly.subplots for making subplots
import plotly.graph_objects as go #Importing graph objects from plotly

#Describing the structure of the plotly figure
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("years of Experience Vs the median Yearly brutto salary (2019)","years of Experience Vs the median Yearly brutto salary (2020)"))

#Describing about the first plot in the plotly figure.
fig.add_trace(go.Scatter(name='Yearly brutto salary 2019',x=yoe_2019, y=ybs_2019),
              row=1, col=1)
fig.update_xaxes(title_text="Current salary 2019 (median)", row=1, col=1)
fig.update_yaxes(title_text="The Yearly brutto salary of the year 2019", row=1, col=1)

#Describing about the second plot in the plotly figure.
fig.add_trace(go.Scatter(name='Yearly brutto salary 2020',x=yoe_2020, y=ybs_2020),
              row=1, col=2)
fig.update_xaxes(title_text="Current salary 2020 (median)", row=1, col=2)
fig.update_yaxes(title_text="The Yearly brutto salary of the year 2020", row=1, col=2)

#Changing the size of the plotly plot rather than it's usual size and assigning some title to it.
fig.update_layout(height=600, width=1500,
                  title_text="The median Yearly brutto salary (without bonus and stocks) Vs Current Salary(median)")

fig.show()
```
